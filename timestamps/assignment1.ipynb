```python
import pandas as pd
import numpy as np
raw_data = pd.read_json("export.json",lines=True, orient='columns',typ='frame')
raw_data.head()
col=raw_data.columns
data=pd.DataFrame()
for i in col:
    data[i]=list(raw_data[i].iloc[0].values())
data.head()
data['appOpenedTimeStamps']=raw_data['appOpenedTimeStamps'][0]
data
```

```python
def convert(data):
    data['createdAt']=pd.to_datetime(data['createdAt'])
    #print("Created at",data['createdAt'])
    data['appLastOpened']=pd.to_datetime(data['appLastOpened'])
    #print("opened at",data['appLastOpened'])
convert(data)
```

```python
#diff b/w time stamps of user activity in a range of 10 
import random

ts=[]
for i in range(20):
    x=list(data['appLastOpened'][:10])
    x.sort()
    ts.append(x)
data['timestamps']=ts
#data
```

```python
#time differencess between consecutive timestamps of appLastOpened
def coef(timestamp):
    diff=[]
    for i in range(len(timestamp)-1):
        diff.append(abs((timestamp[i+1] - timestamp[i])))
    #timedelta is stored 
    #print(diff)
    diff.append(timestamp[0]-timestamp[len(timestamp)-1])#first and last time difference
    return diff
lis=[]
x=coef(list(data['appLastOpened']))
data['TimeDifference_lastopened']=x
for i in data['timestamps']:
    #print(i)
    x=coef(i)
    lis.append(x)
data['Coefficient']=lis
#time differences of appcreated and appLastOpened
def diff_createdandopened(data):
    co=list(zip(data['createdAt'],data['appLastOpened']))
    diff=[]
    for i in range(len(co)):
        diff.append(abs((co[i][0] - co[i][1])))
    return diff
data['diff_createdandopened']=diff_createdandopened(data)
data
```

```python
#assigning high coefficient for least time difference and least for highest respectively
#this is the main logic function
def coeff(timestamp):
    c=range(1,21)
    x=sorted(timestamp)
    x=x[::-1]
    assigncoeff={}
    k=0
    for i in x:
        assigncoeff[i]=c[k]
        k+=1
    coeffs=[]
    for i in timestamp:
        coeffs.append(assigncoeff[i])
    print(coeffs)
    return coeffs
data['coeff_lastOpened']=coeff(data['TimeDifference_lastopened'])
data
```
